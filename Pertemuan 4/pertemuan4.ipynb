Menampilkan dataset

```python
from pandas import DataFrame
import pandas as pd
import numpy as np

# baca file dataset
df = pd.read_csv("Apartemen_ok.csv")
print(df)
df.info()
```

Berikan tambahan kolom dengan nama kolom Label

```python
# membuat list yang akan ditambahkan menjadi kolom tambahan
label_kelas = ['Murah', 'Murah', 'Mahal', 'Murah', 'Mahal', 'Mahal', 'Mahal', 'Murah', 'Murah']

# memberi nama kolom tambahan 'Label'
df['Label'] = label_kelas
print(df)
```

Tampilkan statistik deskriptif (khusus atribut numerik)

```python
# menampilkan statistik deskriptif, hanya numerik
print(df.describe())
```

Hitung data unik pada kolom / atribut Wilayah

```python
# mengeksplorasi lebih dalam untuk antribut wilayah (tipe kategori)
df.Wilayah.unique()
valueCount = df.Wilayah.value_counts()
print(valueCount)
```

Tampilkan diagram batang kolom / atribut Wilayah

```python
# visualisasi dalam diagram batang kolom wilayah
import seaborn as sns

sns.set(style="darkgrid")
ax = sns.countplot(x="Wilayah", data=df)

print(ax)
```

```python
# bentuk lain plot kolom Wilayah (ada text jumlah)
import matplotlib.pyplot as plt # library for plot

plt.figure(figsize=(5, 5))

# total = float(len(df["Label"]))
ax = sns.countplot(x="Wilayah", data=df)
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x() + p.get_width() / 2.,
            height,
            '{:1.0f}'.format((height)),
            ha="center")

plt.show()
```

Tampilkan histogram untuk semua atribut numerik

```python
# menampilkan histogram untuk semua atribut numerik
import matplotlib
import matplotlib.pyplot as plt

numerik_df = df.select_dtypes(include=['float64'])
numerik_index = numerik_df.columns
numerik_index
df.hist(column=numerik_index, figsize=(10, 20), layout=(5, 1))
plt.show()
```
